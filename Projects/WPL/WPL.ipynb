```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("wpl.csv")
```

```python
plt.rcParams['font.family']='Segoe UI'
```

```python
target=df['Runs_1'].sum()+1
fig=plt.figure(figsize=(10,6.5))
fig.set_facecolor('#FADA7A')
ax1=fig.add_subplot()
ax1.set_facecolor('#F4F8D3')

runs_2=df['Runs_2'][:17]
overs_2=df['Overs'][:17]
wickets_2=df['Wkts_2'][:17]
rr_2=df['Required_RR'][:17]

ax1.plot(overs_2,runs_2,marker='o',label='Runs',color="#0f1059")
ax1.plot(overs_2,rr_2,marker='o',label='Required Runrate',color="#c9000d")

ax1.set_xticks(np.arange(1,18))
ax1.set_ylim(0,15)
ax1.set_xlim(1,17)

ax1.fill_between(overs_2,runs_2,rr_2,where=(runs_2>rr_2),alpha=0.7,interpolate=True)
ax1.fill_between(overs_2,runs_2,rr_2,where=(runs_2<rr_2),alpha=0.7,interpolate=True)
ax1.grid(alpha=0.4,color='black',linestyle='--')

ax1.set_title(f"Runs scored vs Required runrate of RCB vs DC while chasing {target}",fontsize=18,fontdict={'fontname':'Century Gothic','weight':'bold'})
ax1.set_xlabel("Overs",fontsize=14,fontdict={'fontname':'Trebuchet MS','weight':'bold'})
ax1.set_ylabel("Runs and Required runrate",fontsize=14,fontdict={'fontname':'Trebuchet MS','weight':'bold'})
ax1.text(8.5,0.3,'@Copyright',color="black",fontsize=12,alpha=0.5)
# Background Color for Overs
ax1.axvspan(1, 7, color='cyan', alpha=0.5, label="Powerplay")  # Light Green
ax1.axvspan(7, 16, color='#ffabe4', alpha=0.5, label="Middle Overs") # Yellow
ax1.axvspan(16, 17, color='#00FF9C', alpha=0.5, label="Death Overs") # Pink

ax1.legend(prop={'family':'Segoe UI',})
# plt.title("hello")
plt.show()
```
