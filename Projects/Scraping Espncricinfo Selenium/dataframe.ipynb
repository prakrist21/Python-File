```python
def perOverCalculation(df):
    detaildf=pd.DataFrame(columns=['Over','Ball','Detail',"Bowler"])
    for m in range(len(df)):
        ball=1
        for y in df['OverDetail'][m].split(','):
            if (y[1:]=='') or (y[1:] in 'Wlb'):
                l=len(detaildf)
                detaildf.loc[l]=[df['Over'][m],ball,y,df['Bowler'][m]]
                print(df['Over'][m],ball,y)
                ball=ball+1
            elif y[1:] in 'wnb':
                l=len(detaildf)
                detaildf.loc[l]=[df['Over'][m],ball,y,df['Bowler'][m]]
                print(df['Over'][m],ball,y)
    return detaildf


def whatevent(row):
    if 'w' in row:
        return "Wide"
    elif 'W' in row:
        return "Wicket"
    elif 'lb' in row:
        return "Leg by"
    elif 'nb' in row:
        return "No ball"
    else:
        return "Runs"

def runsevent(row):
    if 'w' in row:
        runs=int(row[0])
        return runs
    elif 'W' in row:
        try:
            runs=int(row[0])
            return runs
        except:
            return 0
    elif 'lb' in row:
        runs=int(row[0])
        return runs
    elif 'nb' in row:
        runs=int(row[0])
        return runs
    else:
        return int(row[0])
```

```python
t1detail=perOverCalculation(t1)          
t1detail['Event']=t1detail['Detail'].apply(whatevent)
t1detail['Runs']=t1detail['Detail'].apply(runsevent)

t2detail=perOverCalculation(t2)          
t2detail['Event']=t2detail['Detail'].apply(whatevent)
t2detail['Runs']=t2detail['Detail'].apply(runsevent)
```
