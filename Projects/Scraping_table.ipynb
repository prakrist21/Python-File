```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://en.wikipedia.org/wiki/List_of_FIFA_World_Cup_finals"
```

```python
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.text,'html')
```

```python
table=soup.find_all('table')[3]
```

```python
head=table.find_all('th')[:8]
```

```python
header=[x.text.strip() for x in head]
```

```python
df=None
df=pd.DataFrame(columns=header)
```

```python
body=table.find_all('tr')
```

```python
for row in body[1:]:
  indi_row=row.find_all('td')
  indi_th=row.find_all('th')
  indi_row.insert(0,indi_th[0])
  indi_row_data=[x.text.strip() for x in indi_row]
  # print(indi_row_data)
  l=len(df)
  df.loc[l]=indi_row_data
df
```
